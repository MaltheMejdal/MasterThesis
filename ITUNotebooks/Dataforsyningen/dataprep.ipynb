### Renaming tiles so it matches the Coordinate system of Tile2Net

```python
import os

# Define the directory path
directory_path = r'C:\Users\Nielsen\Desktop\tile2net\tile2net-main\ITUData\dataforsyningen\tiles\static\256_15'

# Define the old and new ranges for x and y
old_x_range = range(682027, 682038)
old_y_range = range(400997, 401005)
new_x_range = range(553978, 553989)
new_y_range = range(325709, 325715)

# Iterate over the files in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.png'):
        # Extract x and y from the filename
        x, y = map(int, filename[:-4].split('_'))

        # Check if x and y fall within the old ranges
        if x in old_x_range and y in old_y_range:
            # Calculate the corresponding new x and y
            print(x)
            new_x = new_x_range[old_x_range.index(x)]
            print(y)
            new_y = new_y_range[old_y_range.index(y)]

            # Form the new filename
            new_filename = f"{new_x}_{new_y}.png"

            # Rename the file
            os.rename(os.path.join(directory_path, filename), os.path.join(directory_path, new_filename))
            print(f"Renamed {filename} to {new_filename}")
        else:
            print(f"File {filename} is not within the specified range.")
```
