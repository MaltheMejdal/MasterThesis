```python
import os
from PIL import Image

def resizeTile(input_dir, outputDir, target_size=(256, 256)):
    # Create output directory if it doesnt exist
    if not os.path.exists(outputDir):
        os.makedirs(outputDir)

    # Iterate through all files in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".png"):
            # Open image
            imgPath = os.path.join(input_dir, filename)
            img = Image.open(imgPath)

            # Resize image
            resizedImg = img.resize(target_size)

            # Save resized image to output directory
            outputPath = os.path.join(outputDir, filename)
            resizedImg.save(outputPath)
            print(f"Resized and saved: {outputPath}")


# Input tile directory
input_directory = r"C:\Users\Nielsen\Desktop\tile2net\tile2net-main\ITUData\MapBox\Aarhus1\tiles\static\512_20"

# Output directory to save resized tiles
output_directory = r"C:\Users\Nielsen\Desktop\tile2net\tile2net-main\ITUData\MapBox\Aarhus1\tiles\static\256_20"

#Target tile size
targetDim = (256, 256)

#Resizing tiles. 
resizeTile(input_directory, output_directory, targetDim)
```
